```python
def convert_type(x):
    ty = type(x)
    if ty == str:
        try:
            return int(x)
        except:
            try:
                return float(x)
            except:
                pass
    return x
```

```python
df_final['attribute_value'] = df_final.attribute_value.apply(lambda x: convert_type(x))
```
